# 10_08 박스 플롯

```python
# 필요한 패키지 임포트

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# 데이터 불러오기
df = pd.read_csv('/Users/adam/Data_Analytics/Python/Datasets/50_Startups/50_Startups.csv')

# 데이터 샘플 확인
df.head()
```

```python
# Profit 변수로 기본 가로 세로 박스 플롯 시각화

# 세로 박스 플롯
plt.figure(figsize=(8,6))
sns.boxplot(y='Profit', data=df)
plt.show()

# 가로 박스 플롯
plt.figure(figsize = (8,2))
sns.boxplot(x='Profit', data=df)
plt.show()

# 중앙값이 다소 낮은 위치에 있어서 데이터 분포가 오른쪽으로 치우쳐 있을 것으로 판단된다.
```

```python
# State 구분에 따른 Profit 박스 플롯 시각화

plt.figure(figsize=(8,5))
sns.boxplot(x='State', y='Profit', data=df)
plt.show()
```

```python
# 평균, 데이터 포인트 포함한 박스 플롯 시각화

sns.boxplot(x="State", y="Profit", 
            showmeans=True, 
            boxprops={'facecolor':'None'}, 
            data=df)

sns.stripplot(x='State', y='Profit', 
              data=df, 
              jitter=True, 
              marker='o', 
              alpha=0.5,
              color='black')

plt.show()
```

```python
# 플로리다의 수익이 가장 높고 왼꼬리 분포의 성향이 가장 강할 것으로 판단된다.
```
